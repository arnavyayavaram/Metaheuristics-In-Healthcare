```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'dataset' is your dataset where each row represents an image
# 'selected_pixels' is an array indicating selected pixels (1D array of indices)
# 'index' is the row index representing the image you want to visualize

# Example dataset (replace this with your actual dataset)
# dataset = ... (your dataset here)
# selected_pixels = ... (your array of selected pixel indices)
# index = 0  # Change this index to select a different image

# Example dataset generation (for demonstration purposes)
np.random.seed(42)
dataset = np.random.randint(0, 256, size=(100, 784))  # Generating a dummy dataset of 100 images

index = 0  # Selecting the first image for visualization
selected_pixels = [432, 318, 327, 484, 184, 182, 353, 372, 434, 625, 265, 515, 350, 481 ,518 ,656, 237, 241,595, 289, 244, 130, 154, 439, 541, 321, 324, 149]  # Replace this with your array of selected pixel indices

# Extract the image from the dataset and reshape it to 28x28
image = dataset[index].reshape(28, 28)
selected_image = np.zeros_like(image)
selected_image.flat[selected_pixels] = image.flat[selected_pixels]

# Plot the original and selected images
plt.figure(figsize=(8, 4))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

# Selected Pixels Highlighted Image
plt.subplot(1, 2, 2)
plt.imshow(selected_image, cmap='gray')
plt.title('Selected Pixels Highlighted')

plt.show()
```
